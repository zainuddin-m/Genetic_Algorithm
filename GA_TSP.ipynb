## Problem Description:
How to design a genetic algorithm to solve the Travelling Salesman Problem:
1. Encoding Scheme:
Each solution (or route) is represented as a list of integers where each integer corresponds to a city. For example: [0, 1, 2, 3, ..., 16] represents the order in which the cities are visited.
2. Fitness Function:
The fitness function calculates the total distance of the route.
Add up the distances between consecutive cities in the route.
3. Genetic Operators:
- Selection: Use tournament selection where the best of 3 randomly selected routes is chosen for reproduction.
- Crossover: Partially Matched Crossover (PMX):Split the route into two parts. Combine the first part of one parent with the remaining part of the second parent, ensuring no duplicates.
- Mutation: Randomly swap the positions of two cities in the route to maintain diversity.

```
!pip install deap
import array
import random
import numpy as np
from deap import algorithms, base, creator, tools
```

```
# fitness function as the sum of all the distances between each consecutive cities in individual
def evalTSP(individual):
    distance = distance_map[individual[-1]][individual[0]]
    for gene1, gene2 in zip(individual[0:-1], individual[1:]):
        distance += distance_map[gene1][gene2]
    return distance,

# create creator.FitnessMin and creator.Individual to be used in the toolbox
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", array.array, typecode='i', fitness=creator.FitnessMin)

# create the toolbox from the base
toolbox = base.Toolbox()

# add attribute generator to the toolbox
toolbox.register("indices", random.sample, range(TourSize), TourSize)

# initializers for individuals and population
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# genetic operators and fitness function
toolbox.register("mate", tools.cxPartialyMatched)
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05)
# toolbox.register("select", tools.selRoulette)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evalTSP)
```

```
random.seed(16)
pop = toolbox.population(n=500)
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("min", np.min)
algorithms.eaSimple(pop, toolbox, 0.7, 0.2, ngen=50, stats=stats)
# algorithms.eaSimple(pop, toolbox, 0.7, 0.2, ngen=50, stats=stats, verbose=False)

best = tools.selBest(pop, 1)[0]
print("Best: %s\nTotal Distance: %s" %(best.tolist(), best.fitness.values))
```
